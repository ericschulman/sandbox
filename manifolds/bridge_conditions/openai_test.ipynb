```python
from dotenv import load_dotenv
import os
import sys

def main():
    # Load .env from current working directory (project root)
    ok = load_dotenv()
    print(f"load_dotenv() returned: {ok}")

    api_key = os.getenv("OPENAI_API_KEY")
    if not api_key:
        raise RuntimeError(
            "OPENAI_API_KEY not found. Ensure you have a .env file in this folder "
            "and it contains a line like OPENAI_API_KEY=sk-..."
        )

    # Print only a safe prefix so you can verify it loaded without leaking the key
    print("OPENAI_API_KEY loaded:", api_key[:8] + "...")

    # Optional: show which Python is running (helps debug Anaconda/Jupyter issues)
    print("Python executable:", sys.executable)

    # Test a minimal API call
    from openai import OpenAI
    client = OpenAI(api_key=api_key)

    resp = client.responses.create(
        model="gpt-4.1-mini",
        input="Say 'ok' and nothing else."
    )

    print("API response:", resp.output_text)

if __name__ == "__main__":
    main()
```
